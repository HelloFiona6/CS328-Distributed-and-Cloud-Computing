brute force latency = 0.414592


```python
import matplotlib.pyplot as plt

# 
np_list = [1, 2, 4, 8, 16, 32]
latency = [0.391761, 0.252560, 0.166810, 0.127435, 0.178174, 0.201580]
bf_latency = 0.414592  # brute force latency

# 
plt.figure(figsize=(8, 5))
plt.plot(np_list, latency, marker='o', linestyle='-', color='#845ec2', label='MPI Parallel Latency')
plt.axhline(y=bf_latency, color='#ff6f91', linestyle='--', label=f'Brute Force ({bf_latency:.3f}s)')

# 
plt.title('Parallel Matrix Multiplication Performance (N=500)')
plt.xlabel('Number of Processes (np)')
plt.ylabel('Latency (seconds)')
plt.xticks(np_list)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()

# 
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 
N = [10, 50, 100, 250, 500]
data = {
    1:  [0.000022, 0.000602, 0.005239, 0.083137, 0.391761],
    2:  [0.000173, 0.000502, 0.002807, 0.038978, 0.252560],
    4:  [0.000377, 0.000424, 0.001822, 0.023588, 0.166810],
    8:  [0.000590, 0.000592, 0.002040, 0.020713, 0.127435],
    16: [0.000973, 0.002289, 0.006038, 0.039310, 0.175558],
    32: [0.002276, 0.010291, 0.018191, 0.041004, 0.201580],
    'bf': [0.000004, 0.000569, 0.004276, 0.080938, 0.414592]
}
```

```python
plt.figure(figsize=(9,6))

#  np 
for np_val, latencies in data.items():
    if np_val == 'bf':
        plt.plot(N, latencies, 'r--', marker='o', linewidth=2, label='Brute Force')
        # 
        # for x, y in zip(N, latencies):
        #     plt.text(x, y, f'{y:.6f}', fontsize=9, color='red', ha='left', va='bottom')
    else:
        plt.plot(N, latencies, marker='o', linestyle='-', label=f'np={np_val}')

# 
plt.title('Matrix Multiplication Latency vs Matrix Size (N)')
plt.xlabel('Matrix Size (N)')
plt.ylabel('Latency (seconds)')
plt.xticks(N)  # 
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.tight_layout()
plt.show()
```
